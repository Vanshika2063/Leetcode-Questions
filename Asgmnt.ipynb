```python
#Q1
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
#I: Scatter plot
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]
plt.scatter(x, y)
plt.show()
```

```python
#II: Histogram
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
#III: Piechart
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels)
plt.legend(title = "Fruits")
plt.show()
```

```python
#IV: Bar chart
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = "blue")
plt.show()
```

```python
#V: Box plot
np.random.seed(10)
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data)
plt.show()
```

```python
#VI: Violin plot
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.violinplot(data_to_plot)
plt.show()
```

```python
#VII: Heat map
array = np.random.rand(10, 10)
sns.heatmap(array, linewidth = 0.5, cmap = 'coolwarm')
plt.title("Heat Map")
plt.show()
```

```python
#Q2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
#Dataset 1
dataset = pd.read_csv("Dataset 1.csv", nrows = 1000)
```

```python
#I: Scatter plot
rows = dataset.loc[dataset["model"] == "combo"]
rows.plot.scatter("year", "province", title = "Scatter plot")
```

```python
#II: Histogram
plt.title("Histogram")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.hist(dataset["fuel"])
```

```python
#III: Piechart
x = dataset["fuel"].value_counts()
labels = ["Gasoline", "Diesel", "LPG", "Hybrid", "Electric", "CNG"]
plt.pie(x)
plt.legend(title = "Pie Chart", labels = labels, loc = 1)
plt.show()
```

```python
#IV: Bar chart
x = ["Gasoline", "Diesel", "LPG", "Hybrid", "Electric", "CNG"]
y = (dataset["fuel"].value_counts()).values.tolist()
plt.bar(x, y, color = "blue")
plt.title("Bar chart")
plt.show()
```

```python
#V: Box plot
dataset.boxplot(by = ["model"], column = ["vol_engine"], grid = False)
```

```python
#VI: Violin plot
sns.violinplot(x = dataset["model"], y = dataset["vol_engine"], scale = "width")
plt.title("Violin plot")
plt.show()
```

```python
#VII: Heat map
array = np.asarray(dataset["price"]).reshape(100, 10)
sns.heatmap(array, linewidth = 0.5, cmap = 'coolwarm')
plt.title("Heat Map")
plt.show()
```

```python
#Q3
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
#Dataset 1
dataset = pd.read_csv("Dataset 6.csv")
```

```python
plt.figure(figsize = (13,10), dpi = 80)
group_col = 'Gender'
order_of_bars = dataset.Stage.unique()[::-1]
colors = [plt.cm.Spectral(i/float(len(dataset[group_col].unique())-1)) for i in range(len(dataset[group_col].unique()))]
for c, group in zip(colors, dataset[group_col].unique()):
    sns.barplot(x = 'Users', y = 'Stage', data=dataset.loc[dataset[group_col]==group, :], order = order_of_bars, color = c, label = group)
plt.xlabel("$Users$")
plt.ylabel("Stage of Purchase")
plt.yticks(fontsize = 12)
plt.title("Population Pyramid of the Marketing Funnel", fontsize = 22)
plt.legend()
plt.show()
```
